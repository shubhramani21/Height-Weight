```python
import pandas as pd
import numpy as np
import seaborn as sn
```

```python
df=pd.read_csv("weight-height.csv")
```

```python
df
```

```python
df['BMI']=(df.Weight/(df.Height**2))*703
```

```python
df
```

```python
sn.histplot(df.BMI,kde=True)
```

```python
info=df.BMI.describe()
```

```python
info
```

```python
MinRange=(df.BMI.mean()-3*df.BMI.std()).round(2)
```

```python
MaxRange=(df.BMI.mean()+3*df.BMI.std()).round(2)
```

```python
MinRange
```

```python
MaxRange
```

```python
df[df['BMI']>MaxRange]
```

```python
df[df['BMI']<MinRange]
```

```python
df_outlier=df[(df['BMI']<MaxRange) & (df['BMI']>MinRange)]
```

```python
sn.histplot(df_outlier.BMI,kde=True)
```

```python
df_fit=df[(df['BMI']>18.5) & (df['BMI']<24.9)]
```

```python
sn.histplot(df_fit['BMI'],kde=True)
```
